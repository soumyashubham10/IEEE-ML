
Overview
This project implements an end-to-end semantic search engine using transformer-based sentence embeddings and FAISS. The system retrieves semantically relevant passages for a given query instead of relying on keyword matching.Dataset
- **Dataset:** MS MARCO (v1.1)
- **Source:** Hugging Face Datasets
- **Passages Used:** ~75,000 (subset due to resource constraints)
- **Domain:** Open-domain information retrieval

Tools & Libraries
- Python
- Pandas
- Hugging Face Datasets
- SentenceTransformers
- FAISS
- NumPy
- PyTorch

 Methodology
1. Load MS MARCO dataset using Hugging Face
2. Extract passage texts and create a corpus
3. Generate embeddings using `all-MiniLM-L6-v2`
4. Normalize embeddings for cosine similarity
5. Build FAISS index for fast retrieval
6. Perform semantic search using query embeddings
7. Evaluate results using Precision@K

Evaluation Metrics
- Precision@K
- Recall@K
- Mean Reciprocal Rank (MRR)
